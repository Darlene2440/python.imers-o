```python
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Ensure df is loaded and df_limpo is created
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
df_limpo = df.dropna()

# Calculate average salary by experience level
df_salario_medio_por_nivel = df_limpo.groupby('experience_level')['salary_in_usd'].mean().reset_index()

# Calculate remote work ratio
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

# Create subplots: 2 rows, 2 columns. The histogram will span across columns in the second row.
fig = make_subplots(
    rows=2, cols=2,
    column_widths=[0.6, 0.4],
    row_heights=[0.6, 0.4],
    specs=[[{'type': 'xy'}, {'type': 'domain'}],
           [{'type': 'xy', 'colspan': 2}, None]], # Explicitly define specs as dictionaries
    subplot_titles=("Média Salarial por Nível de Experiência", "Proporção de Tipos de Trabalho Remoto", "Distribuição de Salários Anuais") # Corrected title order
)

# Add Bar Chart (Average Salary by Experience Level)
fig.add_trace(
    go.Bar(
        x=df_salario_medio_por_nivel['experience_level'],
        y=df_salario_medio_por_nivel['salary_in_usd'],
        name='Média Salarial',
        marker_color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Example colors
    ),
    row=1, col=1
)

# Add Pie Chart (Remote Work Ratio)
fig.add_trace(
    go.Pie(
        labels=remoto_contagem['tipo_trabalho'],
        values=remoto_contagem['quantidade'],
        name='Tipo de Trabalho',
        hole=0.3,
        marker_colors=['#8c564b', '#e377c2', '#7f7f7f'] # Example colors
    ),
    row=1, col=2
)

# Add Histogram (Overall Salary Distribution)
fig.add_trace(
    go.Histogram(
        x=df_limpo['salary_in_usd'],
        name='Salário em USD',
        nbinsx=50,
        marker_color='#9467bd'
    ),
    row=2, col=1 # This trace will span two columns due to specs definition
)

# Update layout
fig.update_layout(
    title_text='Dashboard de Análise de Salários',
    height=700,
    showlegend=False
)

# Update x-axis label for bar chart
fig.update_xaxes(title_text='Nível de Experiência', row=1, col=1)
fig.update_yaxes(title_text='Média Salarial (USD)', row=1, col=1)

# Update x-axis label for histogram
fig.update_xaxes(title_text='Salário em USD', row=2, col=1)
fig.update_yaxes(title_text='Frequência', row=2, col=1)

fig.show()
```
